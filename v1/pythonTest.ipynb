```python
import random
r = random.randint(0,100)
print(r)
try:
  x = int(input('請猜一個數字(0-100)：'))
  while(x!=r):
    if(x>100):
      print('數字太大了')
    else:
      print('猜錯了，請在試一次') 
    x = int(input('請猜一個數字(0-100)：'))
    
  print('猜對了')
except ValueError:
  print('發生輸入非數值的錯誤!') 
```

# 第五題

```python
import requests
url = 'https://hispark.hccg.gov.tw/OpenData/GetParkInfo'
data = requests.get(url)
data = data.json()
x=input('請輸入停車場名稱:')
for stitle in data:
  if x in stitle['PARKINGNAME']:
    print('\n'+'顯示其地址：'+stitle['ADDRESS']+'\n'+'平日費率：'+stitle['WEEKDAYS']+'\n'+'假日費率：'+stitle['HOLIDAY']+'\n'+'汽車剩餘車位：'+str(stitle['FREESPACE']))
```

# 第六題

```python
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests

url = 'https://hispark.hccg.gov.tw/OpenData/GetParkInfo'
data = requests.get(url)
data = data.json()
xnum,ynum = [],[]

for stitle in data:
  xnum1 = stitle['PARKNO']
  ynum1 = stitle['FREESPACE']
  xnum.append(xnum1)
  ynum.append(ynum1)

plt.plot(xnum, ynum,'r:o', linewidth=2.0)
plt.title('新竹市剩餘車位資訊')
plt.xlabel('編號')
plt.ylabel('剩餘位置')
plt.show()
```
